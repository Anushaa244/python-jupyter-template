```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]
y_bar = [5, 7, 3, 8, 6]
y_scatter = [2, 3, 5, 7, 11]
sizes = [30, 25, 25, 20]
labels = ['Python', 'Java', 'C++', 'Javascript']

# Check if the data lists have the same length
def check_data_length(*args):
    length = len(args[0])
    for arg in args:
        if len(arg) != length:
            raise ValueError(f"Data lists must all have the same length. Found {len(arg)} instead of {length}.")

# Check if the data lengths are consistent
check_data_length(x, y_line, y_bar, y_scatter)

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(x, y_bar, color='orange')
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y_scatter, color='green')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show all plots
plt.show()
```
