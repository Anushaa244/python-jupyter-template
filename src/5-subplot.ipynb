```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y_line = [2, 3, 5, 7, 11]
y_bar = [5, 7, 3, 8, 6]
y_scatter = [1, 4, 9, 16, 25]
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y_line, marker='o')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(x, y_bar, color='orange')
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y_scatter, color='green')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
