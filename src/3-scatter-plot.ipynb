```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling: Ensure x and y have the same length
if len(x) != len(y):
    raise ValueError("x and y lists must have the same length.")

# Create scatter plot
plt.scatter(x, y, color='purple')

# Annotate each point
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(-5, 5), ha='center')

# Customize the plot
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.grid(True)
plt.show()
```
